#### 5. Identify the categorical and continuous data

```python
def seprate_data_types(cust_data):
    categorical = []
    continuous = []
    for column in cust_data.columns:
        if cust_data[column].nunique() < 100:
            
            categorical.append(column)
        else:
            continuous.append(column)
    
    return categorical, continuous

categorical, continuous = seprate_data_types(cust_data)        

from tabulate import tabulate
table = [categorical, continuous]
print(tabulate({"Categorical":categorical,
                "Contiuous": continuous}, headers = ["categorical", "continuous"]))
```

#### 6. Missing values treatment

```python
gender = cust_data["Gender"].mode()[0]
card = cust_data["Card_Category"].mode()[0]
credit_limit = cust_data["Credit_Limit"].mean()
```

```python
cust_data["Gender"] = cust_data["Gender"].fillna(gender)
cust_data["Card_Category"] = cust_data["Card_Category"].fillna(card)
cust_data["Credit_Limit"] = cust_data["Credit_Limit"].fillna(credit_limit)
```

```python
cust_data.isnull().sum()
```

#### This function will check for the following
##### 1. Unique value
##### 2. Mode
##### 3. Total count of missing values

```python
def info_of_cat(col): 
    print(f"Unique values in {col} are: {cust_data[col].unique()}")  
    print(f"Mode of {col} is {cust_data[col].mode()[0]}")  
    print(f"Number of missing values in {col} is {cust_data[col].isnull().sum()}")
```

#### 14. How is the credit limit of the blue card custbomers affecting the churn?

```python
blue_customers = cust_data[cust_data["Card_Category"] == "Blue"]

# Calculate churn rate for blue card customers
total_customers = len(blue_customers)
churned_customers = len(blue_customers[blue_customers["Attrition_Flag"] == "Attrited Customer"])
churn_rate = churned_customers / total_customers

# Print the churn rate
print("Churn rate for blue card customers: {:.2%}".format(churn_rate))

# Analyze how credit limit affects the churn rate
credit_limit_bins = pd.cut(blue_customers["Credit_Limit"], bins=10)
churn_rates_by_credit_limit = blue_customers.groupby(credit_limit_bins)["Attrition_Flag"].apply(lambda x: len(x[x == "Attrited Customer"]) / len(x)).fillna(0)

# Visualize the results using a line chart
plt.figure(figsize=(12,6))
ax = churn_rates_by_credit_limit.plot.line()
ax.set_xlabel("Credit Limit")
ax.set_ylabel("Churn Rate")
ax.set_title("Churn Rate vs. Credit Limit for Blue Card Customers")
plt.show()
```

##### Interpretation: It is seen that churn rate is higher where credit limit is less than 5000 dollars. but as it is negative correlation till approx 20000 dollars we can conclude as credit limit inceasese customers are less likely to churn. However after 20000 dollars it is positive correlation, it means that customers with higher credit limits are more likely to churn. may be some another reason are there for churn rate.
